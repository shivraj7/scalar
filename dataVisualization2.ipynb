```python
tp3_genre = data['Genre'].value_counts().head(3).index
tp3_genre
```

```python
tp3_platform = data['Platform'].value_counts().head(3).index
tp3_platform
```

```python
#now i want row in which top3 publisher,genre,platform are present
data['Publisher'].isin(tp3_publisher)
```

```python
data[data['Publisher'].isin(tp3_publisher)]
```

```python
tp3_data = data[data['Publisher'].isin(tp3_publisher) & data['Genre'].isin(tp3_genre) & data['Platform'].isin(tp3_platform)]
tp3_data
```

```python
#which is the most common pulblisher? - Electronic Art
sns.countplot(data=tp3_data, x='Publisher')
plt.show()
```

```python
#which is most common publisher and which genre is most common in it- Sports
sns.countplot(data=tp3_data, x='Publisher', hue='Genre')
plt.show()
```

```python
#which is most common publisher and which Platform is most common in it- PS2
sns.countplot(data=tp3_data, x='Publisher', hue='Platform')
plt.show()
```
