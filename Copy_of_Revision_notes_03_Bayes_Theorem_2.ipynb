### <font color='Blue'>**Formulas learnt so far**</font>

Let's recall all the formulas that we have learned so far,  

1. **Conditional probability:**  **$P[A|B] = \Large\frac{P[A∩B]}{P[B]}$**  

2. From conditional probability we will get,

 **$P[A\cap B] = P[A|B] \ * \ P[B]$**  

 which is known as **Multiplication Rule**


3. **Bayes Theorem:** **$P[A|B] = \Large\frac{P[B|A] \ * \ P[A]}{P[B]}$**  

4. **Law of total probability:** **$P(A) = \sum_{i=1}^n P(A \mid B_i)  P(B_i)$**  

5. **Independent Events:** **$P[A\cap B] = P[A] \ * \ P[B] $**  



Now let's verify one claim.

## <font color='blue'>**Claim: If A and B are mutually Exclusive then A and B are not independent.**<font>   


We know that if A and B are mutually exclusive or Disjoint events:  

- <font color='purple'>$A\cap B= \{  \}$ </font>
<br>Note : $A\cap B$  is a null/empty set  as A and B can't occur at the same time    
- So, <font color='purple'>$P(A∩B) = 0$</font>

But in the case of independent events:   

- <font color='purple'>$P(A∩B) = P(A) * P(B) $</font> ( we just saw above)  

<br>

In the case of mutually exclusive events $P(A ∩ B)$ is not equal to $P(A) * P(B)$, as A and B are not independent.  

Therefore, the claim is proven: If A and B are mutually exclusive, then A and B are not independent.

<br>  
***Alternate Method :***
Using the conditional probability formula:

$P(A|B) = \Large\frac{P(A∩B)}{P(B)}$

<font color='purple'>**For Disjoint events:**</font>  

- $P(A∩B) = 0$  
  - So, <font color='purple'>$P(A|B) = \Large\frac{0}{P(B)} = 0$</font>  

<font color='purple'>**For independent Events:**</font>  

-  $P(A∩B) = P(A) * P(B)$  

  - So, <font color='purple'>$P(A|B) = \Large\frac{P(A) \ * \ P(B)}{P(B)} = P(A)$</font>  

<br>  

As we can see in both the events $P(A|B)$ is different

Hence, we can conclude that :

**If A and B are mutually Exclusive then A and B are not independent.**


## <font color='purple'> **Example: 1** </font>

```
In a university, 30% of faculty members are females. Of the female faculty members, 60% have a PHD. Of the male faculty members, 40% have a PHD
- What is the probability that a randomly chosen faculty member is a female and has PHD?
- What is the probability that a randomly chosen faculty member is a male and has PHD?
- What is the probability that a randomly chosen faculty member has a PHD?
- What is the probability that a randomly chosen PHD holder is female?
```

**Explanation:**  

Given,   
- Female faculty members = 30%  
  - Out of this 30% members, 60% have PHD  

- Male faculty members = 100 - 30 = 70%  
  - Out of this 70% members, 40% have PHD  

<br>

**Let's define probabilities:**  

- probability that a randomly chosen faculty member is a female i.e. $P(F) = 0.3$
  - Given that faculty member is a Female, the probability that she has a PHD is i.e $P(phd \ |\ F) = 0.6$
- probability that a randomly chosen faculty member is a Male i.e. $P(M) = 0.7$  
  - Given that faculty member is a Male, the probability that he has a PHD is i.e $P(phd \ |\ M) = 0.4$  

<br>  

**Answering questions:**  

> <font color='purple'>**Q1. What is the probability that a randomly chosen faculty member is a female and has PHD?**  </font>

We know **`AND`** means intersection, here we want to find $P(phd \ ∩ \ F)$  

- Using the formula of conditional probability,  

    $P(phd \ |\ F) = \frac{P(phd \ ∩ \ F)}{P(F)}$  

    So, $P(phd \ ∩ \ F) = P(phd \ |\ F) \ * \ P(F)$    

    Adding values into the equation  

  - $P(phd \ ∩ \ F) = 0.6 * 0.3 = 0.18$  

**Conclusion:**  

<font color='purple'>The probability that a randomly chosen faculty member is a female and has PHD is **0.18**</font>

Similarly,  

<br>  

> <font color='purple'>**Q2. What is the probability that a randomly chosen faculty member is a male and has PHD?**  </font>

- Using the formula of conditional probability,  

    $P(phd \ |\ M) = \frac{P(phd \ ∩ \ M)}{P(M)}$  

    so, $P(phd \ ∩ \ M) = P(phd \ |\ M) \ * \ P(M)$    

    Adding values into the equation  

  - $P(phd \ ∩ \ M) = 0.4 * 0.7 = 0.28$    

**Conclusion:**  

<font color='purple'>The probability that a randomly chosen faculty member is a male and has PHD is **0.28**</font>  

<br>


> <font color='purple'>**Q3. What is the probability that a randomly chosen faculty member has a PHD?**  

We have 2 approaches to solve this question.

<br>

<font color='purple'>**Approach 1:**</font>  

- Here, we need to find the probability that If I choose a random person, then he/she have a PHD, no matter whether the person is MALE or FEMALE.  i.e. $P(phd)$  


- We can add **$P(phd \ ∩ \ F) + P(phd \ ∩ \ M)$** as it'll give me $P(phd)$    

- $P(phd) = P(phd \ ∩ \ F) + P(phd \ ∩ \ M)$  
  
  adding values into the equation  

  - $P(phd) = 0.18 + 0.28 = 0.46$  

<br>  

<font color='purple'>**Approach 2:**</font>

- As we know, we can write **$P(phd \ ∩ \ F)$ as a  $P(phd \ |\ F) \ * \ P(F)$**   because,  $P(phd \ |\ F) = \frac{P(phd \ ∩ \ F)}{P(F)}$  

  Here comes the <font color='purple'>Law of total probability in picture </font>

<br>

- For Male also, we can write <font color='purple'>**$P(phd \ ∩ \ M)$ as a  $P(phd \ |\ M) \ * \ P(M)$**  

<br>

Replacing these values in the equation,  

- $P(phd) = [P(phd \ |\ F) \ * \ P(F)] + [P(phd \ |\ M) \ * \ P(M)] $  
  - $P(phd) = [0.6 * 0.3] + [0.4 * 0.7]$  

   = $P(phd) = 0.46$  

<font color='orange'>**Conclusion:**</font>  

<font color='purple'>The probability that a randomly chosen faculty member has a PHD is **0.46**</font>

<br>  

> <font color='purple'>**Q4. What is the probability that a randomly chosen PHD holder is female?** </font>  

Here, we are already given that the randomly chosen person is PHD holder and we need to find the probability of this person being Female. We need to find: $P(F\ | \ phd)$  

<br>

Using the **formula of conditional probability:**  
- $P(F\ | \ phd) = \frac{P(phd \ \cap \ F)}{P(phd)}$  

  Replace the $P(phd \ \cap \ F)$ with $P(phd \ |\ F) \ * \ P(F)$ ,

  <br>

  and $P(phd)$ with $[P(phd \ |\ F) \ * \ P(F)] + [P(phd \ |\ M) \ * \ P(M)]$  

<br>

Final forumla will be:  

- $P(F\ | \ phd) = \Large\frac{P(phd \ |\ F) \ * \ P(F)}{[P(phd \ |\ F) \ * \ P(F)] \ + \ [P(phd \ |\ M) \ * \ P(M)]}$  

  - $P(F\ | \ phd) = \Large\frac{0.6 * 0.3}{[0.6 * 0.3] + [0.4 * 0.7]}$  

  - $P(F\ | \ phd) = 0.39$  

<font color='orange'>**Conclusion:**</font>  

<font color='purple'>The probability that a randomly chosen PHD holder is female is **0.39**</font>
<br>

> <font color='purple'>**Q. But how much rainfall index is considered as a heavy ranifall?**</font>  

One of the parameter is using the **Median** values of these columns.  

If their individual **rainfall index value > median value** then it'll we considered as **heavy rainfall** and vice a versa  


````python
# new dataset containing only impactful columns

data = df[impactful_columns]

data.head()
````

````python
# let's calculate the median of columns and set as their threshold value

threshold_jun = data['JUN'].median().astype(int)
threshold_jul = data['JUL'].median().astype(int)
threshold_oct = data['OCT'].median().astype(int)
threshold_ar = data['ANNUAL_RAINFALL'].median().astype(int)

threshold_jun, threshold_jul, threshold_oct, threshold_ar
````

````python
thresholds = {
    'JUN': 625,
    'JUL': 691,
    'OCT': 284,
    'ANNUAL_RAINFALL': 2934
}

# Convert columns to binary based on thresholds
for col, threshold in thresholds.items():
    data[col] = (data[col] > threshold).astype(int)

data.head()
````

### <font color='purple'>**Q1. Calculate the Probability of flood given that rainfall in June is greater than the median june rainfall value (threshold for heavy rainfall)**</font>



**Question Explanation**:  


Let A represents : Flood  
B represents: heavy rain in June  

We need to calculate $P(A | B)$ i.e. $\frac{P(A ∩ B)}{P(B)}$  

<br>  

  




#### <font color='purple'>**Solution Approach 1:**</font>  


We can obtain these values using contingency table and put those values into the formula.   

Here we need to compare "FLOODS" and "JUN" column.

````python
pd.crosstab(data['JUN'],
            data['FLOODS'],
            margins=True,
            margins_name='Total')
````

Now, $P(A ∩ B)$ = Probability of Flood occuring `AND` heavy rainfall in JUNE  

As we know in the contingency table, FLOODS = YES represents that flood has occured and JUN = 1 means heavy rainfall.  

We need to check value where FLOODS = YES and JUN = 1 which is **44**    

Then by the formula of condititonal probability we can feed this data




````python
# probability of high rainfall in June P(J)
# P(J) = possible outcomes in june having heavy rainfall / total outcomes

P_J = (16+44)/(42+16+16+44)

# now, P(A and B) (Flood = YES and Jun = 1)

P_F_and_J = 44/(42+16+16+44)

#, so our probability of flood occurring given that the high rainfall occured in June will be

P_F_J = P_F_and_J / P_J

print(f'P(J) : {P_J}')
print(f'P(F AND J) : {P_F_and_J}')
print(f'P(F|J): {P_F_J}')
````

#### <font color='purple'>**Approach 2: using normalize attribute**</font>  


**Explanation of Normalize attribute:**  


Rather putting all the values in the formula and then calculate the probability  

We can just pass one <font color='purple'>more attribute in pd.crosstab()</font> function which will divide all values by the sum of values.   

- This is the probability only, as in probability we divide `possible outcome / total outcome (sum of all values)`  

Parameter is : <font color='orange'>**normalize = ' '**</font>  

<br>

- <font color='purple'>Without this attribute</font>, the contingency table will <font color='purple'>show the raw counts of occurrences for each combination of variables.</font>   

- It will not be normalized, and the values in the table will represent counts.

<br>

Here we can pass these strings in this attribute:

<font color='orange'>**normalize='index'**</font> or <font color='orange'>**normalize='columns'**</font> :   

- The normalize attribute specifies how the values in the contingency table should be normalized.   
  - When set to **'index'**, it <font color='purple'>calculates conditional probabilities  based on rows</font>, treating each row as a separate condition.   

  - When set to **'columns'**, it <font color='purple'>calculates conditional probabilities based on columns</font>, treating each column as the condition we are focusing on.  

- This means that each row in the table is divided by the sum of its row, making each row's values sum up to 1, representing conditional probabilities.  

Same with the column

<br>  



**In this case:**  

By setting <font color='orange'>**normalize='index'**</font>,   

- the code calculates conditional probabilities within each row.   

- Each value in the table represents the probability of the corresponding event (FLOODS) given the value of 'JUN' in that row.   

The row sums up to 1, ensuring that it reflects the conditional probabilities.   

<br>

In summary,   

<font color='purple'>**setting normalize='index'** in pd.crosstab allows you to calculate and visualize conditional probabilities based on the specified row variable ('JUN' in this case)</font>,  

making it easier to assess the impact of one variable on another.

````python
pd.crosstab(index = data['JUN'],
            columns = data['FLOODS'],
            margins=True,
            normalize='index')
````



The values in the table represent the conditional probabilities, where each cell contains the probability of the corresponding outcome (FLOODS) given the condition in June (JUN).  


> <font color='purple'>Then the probability of flood occurring given that the heavy rainfall occured in June will be:</font>

- In the cell at row 1, column 1, the value <font color='orange'>**0.73333**</font> represents the conditional probability of flooding (FLOODS = YES) given that high rainfall occurred in June (JUN = 1).  

#### <font color='orange'>Conclusion:</font>


      So, there is 73.33% chance of Floods when there is a heavy rainfall in June

As we can see by calculating using formula also, we are getting the same answer as using directly conditional probability using normalize = 'index'
