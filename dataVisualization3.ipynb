```python
data = pd.read_csv('final_vg.csv')
```

```python
data
```

Multivariate Analysis


```python
top3_pub = data['Publisher'].value_counts().index[:3]
top3_gen = data['Genre'].value_counts().index[:3]
top3_plat = data['Platform'].value_counts().index[:3]
top3_data = data.loc[(data["Publisher"].isin(top3_pub)) & (data["Platform"].isin(top3_plat)) & (data['Genre'].isin(top3_gen))]
top3_data
```

```python
sns.scatterplot(data=top3_data,x='NA_Sales',y='EU_Sales')
plt.show()
```

```python
sns.scatterplot(data=top3_data,x='NA_Sales',y='EU_Sales', hue='Publisher') # NNC
plt.show()
```

observations

EA sales > activision sales > Namco sales

for both EU and NA


```python
sns.scatterplot(data=top3_data,x='NA_Sales',y='EU_Sales', hue='Genre') # NNC
plt.show()
```

Observations

sports sales > action sales > misc sales

for both EU and NA
