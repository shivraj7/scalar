```python
import pandas as pd
import numpy as np
data = pd.read_csv('final_vg.csv')
data.head()
```

```python
#line plot
x_val = [0,1,2]
y_val = [3,5,9]
plt.plot(x_val,y_val)
```

```python
#how to choose correct plot
# 1 varabble -> univirate
# 2 variable -> bivariate
# >2 variable - > multivariate

#univariate -- numeric data
#           -- categoric data

#bivariate -- numerical, categorical
#          -- categorical categorical
#          -- categorical, numerical

#mutivariate -- categorical, categoric, categoric
#             ...
#             ...

```

```python
plt.figure(figsize=(12,8))
plt.bar(x_val,y_val, color='green')
plt.xticks(rotation =45, fontsize=12)
```

```python
plt.figure(figsize=(10,8))
plt.bar(x_val,y_val,width=0.2,color='orange')
plt.title('Games per Genre',fontsize=15)
plt.xlabel('Genre',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation = 90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
data['Genre'].value_counts().index
```

```python
sns.countplot(x = 'Genre', data = data, order=data['Genre'].value_counts().index, color='cornflowerblue')
plt.xticks(rotation=90)
```

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
region_sales = sales_data.T.sum(axis='columns')

plt.pie(region_sales,
       labels=region_sales.index,
       startangle=90,
       explode=(0.2,0,0,0),
       autopct = '%.2f%%') # label the wedges with their numeric value
plt.show()
```

###Box Plot

```python
plt.figure(figsize=(15,8))
sns.boxplot(y = data["Global_Sales"])
plt.yticks(fontsize=20)
plt.ylabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
```

The above plot is also called box and whisker plot

the black horizontal line are called whisker

the black line whithin blue box is median of value
