#Assignment Q1 Short State

A country has a population average height of 65 inches with a standard deviation of 2.5. A person feels people from his state are shorter. He takes the average of 20 people and sees that it is 64.5.

At a 5% significance level (or 95% confidence level), can we conclude that people from his state are shorter, using the Z-test? What is the p-value?


```python
import numpy as np
from scipy import stats
from statsmodels.stats import power
from scipy.stats import norm

# H0 = Height of the state is same as country
# H1 = Height of the is less than country.

mean_population = 65
sd_population = 2.5

mean_calculated = 64.5
size_sample = 20

significance_level = 0.05
confidance_level = 1 - significance_level

# Use of Z test.
# Step 1. Calculate sample mean and sample variance using CLT
mean_sample = mean_population
sd_sample = sd_population/np.sqrt(size_sample)
print('sample sd : ', sd_sample)
# Step 2. Calculate Z factor for calculated mean.
z = (mean_calculated - mean_sample)/sd_sample
print('z factor : ', z)

# Step 3. Calculate cdf for given z
norm.cdf(z)
print('norm.cdf(z) : ', norm.cdf(z))

# Step 4. Calculate p value
p = norm.cdf(z) # note (1- norm.cdf(z) is p for right tailed test)
print('p : ',p)

if p < significance_level:
  print('Reject hypothesis :- State height is less than country height')
else:
  print('Accept hypothesis :- State height is same as country height')
```
